```python
import pandas as pd
```

```python
ds_ins = pd.read_csv("insurance_pre.csv")
```

```python
ds_ins = pd.get_dummies(ds_ins, drop_first= True)
ds_ins
```

```python
ds_ins.columns
```

```python
ind_x = ds_ins[['age', 'children', 'bmi', 'sex_male', 'smoker_yes']]
dep_y = ds_ins[['charges']]
#ind_x, dep_y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ind_x, dep_y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'], #The functions to be used
              'max_features' : ['sqrt', 'log2'], #The number of features to consider when looking for the best split
              'max_depth': [None, 5, 10, 15],      # Limits tree depth
              'n_estimators':[10,100], #no of trees in the forest
              'random_state' : [0] 
    }

#bind the parameter with the Regression Function
grid_model_ins = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3, n_jobs = -1, cv = 5)
#cv=5: Uses 5-fold cross-validation for better generalization.
#n_jobs=-1: Utilizes all CPU cores to speed up computations.
#verbose=3: Displays detailed logs during execution.
#refit=True: Automatically retrains the model with the best parameters.

#model_ins = RandomForestRegressor(criterion='poisson', splitter='random', random_state = 0)
#model_ins = model_ins.fit(x_train, y_train)
grid_model_ins.fit(x_train, y_train) #fit the model
```

```python
print(grid_model_ins.best_params_)
re = grid_model_ins.cv_results_
#print(re)
```

```python
grid_y_pred = grid_model_ins.predict(x_test)
```

```python
#import matplotlib.pyplot as plt
#from sklearn import tree
#tree.plot_tree(grid_model_ins)
#plt.show()
```

```python
from sklearn.metrics import r2_score
r2_value = r2_score(y_test,grid_y_pred)
r2_value
```
