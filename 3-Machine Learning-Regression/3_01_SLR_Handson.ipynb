```python
# import the libraries
import numpy as np # numpy is used for numerical and Scientifical calculation
import matplotlib.pyplot as plt # import the class pyplot from library matplotlib to handle the data visualization
import pandas as pd # import the library to handle the data
```

```python
#import the data from csv file
data = pd.read_csv('Salary_Data.csv')
```

```python
data
```

```python
x = data[['YearsExperience']] #Independent Variable
y = data[['Salary']] #Dependent Variable
```

```python
plt.scatter(x,y) #assign the data to scatter plot
plt.xlabel('YOE') #label the x axis
plt.ylabel('Sal') #label the y axis
plt.show() #display the scatter plot
```

```python
# Model Selection and split the variable
from sklearn.model_selection import train_test_split # import the function train_test_split from the class model_selection from the library sklearn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0) #assign the splitted data into different variable
```

```python
x_test, y_test
```

```python
from sklearn.linear_model import LinearRegression #import the LinearRegression function from te class linear_model from the library sklearn
regressor = LinearRegression() # create object
regressor.fit(x_train, y_train) # y = mx + c or y = wx1 + bo 
```

```python
#view b1 and bo value (weight(m) and bais(c))
weight = regressor.coef_ # assign the coefficent (slope - m) value into the weight variable
print("Weight of the Model = {}".format(weight))
bais = regressor.intercept_ # assign the intercept (C) value to the bais
print("Intercept of the model={}".format(bais))
```

```python
y_pred = regressor.predict(x_test) # calculate the predicted value for the test set
y_pred
```

```python
# Evaluation metric
from sklearn.metrics import r2_score #import the r Sqr function from metrics class from sklearn lib 
r_score = r2_score(y_test, y_pred) #(SSR / SST); SSR -> Sum of Sqr of (Dep. Predicted - Mean of Dep.); SST -> Sum of Sqr of (Dep. Actual - Mean of Dep.)
r_score # if the value is near to 1 then the model is good
```

```python
import pickle #import pickle library to store and read the model
```

```python
filename = 'finalized_model.sav'
pickle.dump(regressor, open(filename, 'wb')) #Store the model 
```

```python
loaded_model = pickle.load(open('finalized_model.sav', 'rb')) #Read the Model
result = loaded_model.predict([[15]]) #Predict the Dependent Value 
print(result)
```
